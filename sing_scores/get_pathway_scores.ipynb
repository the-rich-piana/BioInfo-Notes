## Get Pathways scores

This notebook will take the expression values from the MUON tutorial and get their pathway scores.

```r
require("GSVA")
require("singscore")
require("GSEABase")
require("data.table")
require(SummarizedExperiment)

# set up the directories
rna_dir = "data/rna_data/"
```

```r

#https://www.celltypist.org/
pathway_score_Bcells = c("RACK1", "HSPA1B", "PTMA", "HBG2", "CD79A", "HLA-DRB1", "RPL10", "MALAT1", "ATP5F1E", "HLA-DRA")
pathway_score_NKcells = c("NKG7","CST3","GNLY","SPINK2","HMOX1","MALAT1","IGFBP4","HBG2","STMN1","KLHL23")
pathway_score_Mcells = c("TYROBP", "FTL", "IGKC", "TRAC", "RPS17", "SAT1", "NEAT1", "LYZ", "HLA-DPB1", "IGLC2")
```

```r
#FOR NK Cells
# read in the gene expression
expr_df = data.frame(fread(paste0(rna_dir, "matrix_rna_nkcells.tsv")), check.names = F)

expr_df[1:10,1:10]

dim(expr_df)

# make the summarized experiment
row.names(expr_df) = expr_df[,1]
expr_df = expr_df[,2:ncol(expr_df)]

expr_se = SummarizedExperiment(assays=list(counts=expr_df[,2:ncol(expr_df)]))
expr_se

# now do the 2 calls needed for singscore
rankData <- rankGenes(expr_se)
#rankData[1:100,1:100]

scoredf <- simpleScore(rankData, upSet = pathway_score_Bcells)
head(scoredf)

scoredf$cell_ids = row.names(scoredf)
scoredf = scoredf[,c("cell_ids", "TotalScore")]
colnames(scoredf)[2] = "NK_cell_score"
write.table(scoredf, paste0(rna_dir, "/sing_scores_onlynk.tsv"), sep="\t", quote=F, row.names=F)
```
